```python
# Amsterdam'da ortalama ev fiyatı nedir?
ortalama_ev_fiyati = housing_data["Price"].mean()
ortalama_ev_fiyati
```

```python
# Amsterdam'daki evlerin medyan fiyatı nedir?
medyan_ev_fiyat = housing_data["Price"].median()
medyan_ev_fiyat
```

```python
# Evlerin ortalama metrekaresi nedir?
ortalama_metrekare = housing_data["Area"].mean()
ortalama_metrekare
```

```python
# Tabloda toplam kaç tane ev listelenmiştir?
toplam_ev_sayisi = housing_data.shape[0]
toplam_ev_sayisi
```

```python
# Minimum ve maximum oda sayısı nedir?
oda_sayisi_araligi = (housing_data["Room"].min(), housing_data["Room"].max())
oda_sayisi_araligi
```

```python
oda_sayisi_araligi[0]
```

```python
oda_sayisi_araligi[1]
```

```python
# Maximum ev fiyatı nedir?
maks_ev_fiyatı = housing_data["Price"].max()
maks_ev_fiyatı
```

```python
# Minimum ev fiyatı nedir? 
min_ev_fiyatı = housing_data["Price"].min()
min_ev_fiyatı
```

```python
# Minimum ev fiyatına sahip olan ev hangisidir?
housing_data[housing_data["Price"] == min_ev_fiyatı]
```

```python
# Maksimum ev fiyatına sahip olan ev hangisidir?
housing_data.query("Price == @maks_ev_fiyatı")
```

```python
# Ortalama fiyat civarında olan evlerin listesi (sıralı) nedir?

housing_data[(housing_data['Price'] < 650000) & (housing_data['Price'] > 600000 )].sort_values(by="Price")
```

## Veriyi görselleştirme gerektiren birkaç soru... 

```python
import seaborn as sns
```

```python
# Amsterdam'daki evlerin fiyat dağılımı ve yoğunluğu
plt.figure(figsize=(10,10))

sns.histplot(housing_data["Price"], bins=100, kde=True, color="green")
plt.title("Ev Fiyat Dağılımı - Amsterdam Agustos, 21")
plt.xlabel("Fiyat")
plt.ylabel("Frekans")
plt.axvline(housing_data["Price"].mean(), color="red")
plt.axvline(housing_data["Price"].median(), color="blue")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,10))

sorted_df = housing_data.sort_values(by="Price")
sorted_df.reset_index(drop=True, inplace=True)
sorted_df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
sorted_df
```

```python
sorted_df.query("Price == 625000")
```

```python
sorted_df.query("Price < @medyan_ev_fiyat +5000 and Price > @medyan_ev_fiyat -5000")
```

```python
sorted_df["Segment"] = "Bilinmiyor"
```

https://stackoverflow.com/questions/13842088/set-value-for-particular-cell-in-pandas-dataframe-using-index

```python
# df.loc[df[<some_column_name>] == <condition>, [<another_column_name>]] = <value_to_add>
sorted_df.loc[sorted_df["Price"] <= 1000000, ["Segment"]] = "1M ve Altı"
```

```python
sorted_df
```

```python
sorted_df.loc[(sorted_df["Price"] > 1000000) & (sorted_df["Price"] <= 2000000), ["Segment"]] = "1M - 2M Arası"
```

```python
sorted_df[800:850]
```

```python
sorted_df.loc[sorted_df["Price"] > 2000000, ["Segment"]] = "2M Üstü"
```

```python
sorted_df
```

```python
# Fiyatlarına göre evleri segmentlere ayırdık, 
# şimdi pie chart ile ev fiyatları segmentlerinin oranını görmek istiyoruz. 

segment_counts = sorted_df['Segment'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(segment_counts, labels=segment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Segmentlere Göre Ev Fiyatları Dağılımı")
plt.legend()
plt.show()
```
